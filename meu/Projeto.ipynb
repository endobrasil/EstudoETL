```python
schema=pa.DataFrameSchema(
    columns={
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_cidade":pa.Column(pa.String),
#"ocorrencia_uf":pa.Column(pa.String,pa.Check.str_length(2,2)),
        "ocorrencia_uf":pa.Column(pa.String),
        "ocorrencia_aerodromo":pa.Column(pa.String),
#"ocorrencia_dia":pa.Column(pa.DateTime), era pra ser isso né, mas tive problemas com date
        "ocorrencia_dia":pa.Column(pa.String),
        "ocorrencia_hora":pa.Column(pa.String, nullable=True),
        "total_recomendacoes":pa.Column(pa.String, nullable=True)      
#"total_recomendacoes":pa.Column(pa.String, required=False) pode ser que um dado específico não tenha esta coluna, aqui ele dispensa ela
        
    }
)
```
